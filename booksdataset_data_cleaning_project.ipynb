## 3. Estandarización y Manejo de los NaN/NaT
**Standardization and Handling of NaN/NaT**

```python
# ========== Handling Price Column ==========
# Notemos que las entradas del campo "Price" presentan el texto "Price Starting at"
# Note that the entries in the "Price" field contain the text "Price Starting at"
df["Price"].head(10)
```

```python
# To-do List: 
# 1) Verificar si las todos los registros presentan el texto "Price Starting at"
# 2) Eliminar texto y modificar nombre del campo por "Price Starting at" 

# To-do List:
# 1) Verify if all records contain the text "Price Starting at"
# 2) Remove the text and rename the field to "Price Starting at"

# Lista para verificar que cada entrada poseea el texto "Price Starting at"
# List to verify that each entry contains the text "Price Starting at"
texto = []

# Lista para almacenar los tipos de moneda registrados
# List to store the recorded currency types
moneda = [] 

# Lista para transformar los valores de string a float
# List to convert string values to floats
valores = [] 

# Los primeros 17 caracteres de cada entrada corresponden a "Price Starting at"
# The first 17 characters of each entry correspond to "Price Starting at"
for registro in df["Price"]:
    texto.append(registro[:17])
    moneda.append(registro[18])
    valores.append(float("".join(registro[19:].split(","))))
```

```python
# Todos los registros poseen el texto "Price Starting at"
# All records contain the text "Price Starting at"
print(f">Todas los registros con 'Price Starting at': {len(texto) == len(df)}")

# Los precios se expresan solo en una moneda
# All prices are listed in one currency only
print(f">Monedas Registradas: {set(moneda)}")
```

```python
# Se agrega la columna/campo "Price Starting At" usando la lista "valores"
# Adding the "Price Starting At" column using the "valores" list
df["Price Starting At"] = valores

# Se elimina la column/campo "Price"
# Removing the "Price" column
df.drop("Price", axis=1, inplace=True)
```

```python
# ========== Handling NaN ==========
# Description: NaN -> Not available
df["Description"] = df["Description"].fillna("Not available")

# Category: NaN - > Unassigned
df["Category"] = df["Category"].fillna("Unassigned")

# Publisher: NaN - > Unknown
df["Publisher"] = df["Publisher"].fillna("Unknown")
```

```python
df.isnull().sum().to_dict()
```

**Observación 3:**

- Se verifica que no existen _NaNs_ en el dataframe tras la modificación.

**Remark 3:**

- No _NaNs_ remain in the dataframe after processing.

```python
# ========== Handling Dates ==========
# Se realiza la siguiente transformación:
# "DayName, MonthName DayNumber, Year" -> YYYY-MM-DD

# Applying the following transformation:
# "DayName, MonthName DayNumber, Year" -> YYYY-MM-DD

def analisis_flexible(fecha_string):
    """
    Utiliza un analizador (parser) inteligente para interpretar 
    de manera flexible el formato de la fecha y ejecutar la 
    transformación al formato YYYY-MM-DD.
    """
    try:
        return parser.parse(fecha_string).date()
    except:
        return pd.NaT
        
# Se crea la columna "Published on" a partir de la aplicación de función anterior
# Creating the "Published on" column from the previously applied function
df["Published on"] = pd.to_datetime(df["Publish Date"].apply(analisis_flexible), errors="coerce")
```

```python
df.info()
```

**Observación 4:**

- Tras el cambio de formato, en la columna "Published on" aparecen 5 registros nulos.


**Remark 4:**

- After reformatting, there are 5 null entries in the "Published on" column.

```python
# Inspección visual de los registros nulos en "Published on"
# Visually inspecting null entries in "Published on"
print(df[df["Published on"].isna()])
```

**Observación 5:**

- Se observa que los 5 registros nulos contienen la fecha _Friday, January 1, 9999_ lo que parece ser un valor por defecto ante la ausencia de la información correspondiente.
- En este caso conviene reemplazar los NaT en la columna "Published on" por la fecha **9999-01-01** como valor por defecto.

**Remark 5:**

- It is observed that the 5 null records contain the date _Friday, January 1, 9999_, which appears to be a default value used when the corresponding information is missing.
- In this case, it is advisable to replace the NaTs in the "Published on" column with the date **9999-01-01** as a default value.

```python
# Se refina el rango de fechas para evitar errores nuevos
# Refining the date range to prevent new errors
df["Published on"] = df["Published on"].astype("datetime64[us]")

# Reemplazo de los NaT con la fecha 9999-01-01
# Replacing NaTs with the date 9999-01-01
df["Published on"] = df["Published on"].fillna(pd.Timestamp("9999-01-01"))
```

```python
# Se verifica que la cantidad de NaT en "Published on" sea cero
# Checking for zero NaT values in the "Published on"
df["Published on"].isnull().sum()
```

```python
# Se elimina la columna "Publish Date"
# Dropping the "Publish Date" column
df.drop(columns=["Publish Date"], inplace=True)

# Chequeo visual del estado final del dataset limpio
# Final visual inspection of the clean dataset
df.head(3)
```

```python
# Se exporta el dataset limpio en formato csv
# Saving the clean dataset as a CSV file
df.to_csv("books-dataset-clean.csv", index=False, encoding="utf-8")
```
